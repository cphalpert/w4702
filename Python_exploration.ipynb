```python
import pandas as pd
import sqlite3
from IPython.display import display, HTML
```

## Table names

```python
# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("events10.sqlite3")
#cur = con.cursor()
table_names = pd.read_sql_query("""SELECT name FROM sqlite_master WHERE type='table';""", con)
table_names
```

## Table sizes

```python
for i, r in table_names.iterrows():
    display(HTML("<h3>%s</h3>" % r['name']))
    display(pd.read_sql_query("""SELECT count(1) FROM %s limit 10;""" % r['name'], con))
```

## Sample for each table

```python
for i, r in table_names.iterrows():
    display(HTML("<h3>%s</h3>" % r['name']))
    display(pd.read_sql_query("""SELECT * FROM %s limit 20;""" % r['name'], con).T)
```

## Event catalog for a specific event
- Looks like a snapshot by request_time

```python
display(pd.read_sql_query("""SELECT * from event_catalog ec
where ec.eventId = 9286407""", con))
```

## Listings for a specific event

```python
display(pd.read_sql_query("""SELECT * FROM listings l
where eventId = 9286407
limit 10""", con).T)
```

## Duplicate entries per listing for a specific event

```python
display(pd.read_sql_query("""SELECT listingId, count(1) FROM listings l
where eventId = 9286407
group by 1""", con))
```

## Listings for a specific eventId & listingId

```python
display(pd.read_sql_query("""SELECT * FROM listings l
where eventId = 9286407 and listingId = 1144268295""", con).T)
```

## Distinct listings per event

```python
display(pd.read_sql_query("""SELECT l.eventId, count(distinct l.listingId) FROM listings l
join events e on e.eventId = l.eventid
group by l.eventId;""", con))
```

## Number of distinct listings

```python
pd.read_sql_query("""SELECT count(distinct l.listingId) FROM listings l""", con).sum()
```

```python
pd.read_sql_query("""SELECT count(l.listingId) FROM listings l""", con).sum()
```
